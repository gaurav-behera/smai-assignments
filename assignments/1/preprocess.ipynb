```python
import numpy as np
import pandas as pd
import os
```

```python
repo_dir = "../../"
raw_data = pd.read_csv(os.path.join(repo_dir, "data", "external", "spotify.csv"), index_col=0)
data = raw_data.copy()
data
```

```python
# Null values handling
# as only the string values are null, we can replace them with "unknown"
data = data.fillna("unknown")
```

```python
# linear normailzation of numerical attributes
data['popularity'] = (data['popularity'] - data['popularity'].min()) / (data['popularity'].max() - data['popularity'].min())
data['duration_ms'] = (data['duration_ms'] - data['duration_ms'].min()) / (data['duration_ms'].max() - data['duration_ms'].min())
data['energy'] = (data['energy'] - data['energy'].min()) / (data['energy'].max() - data['energy'].min())
data['key'] = (data['key'] - data['key'].min()) / (data['key'].max() - data['key'].min())
data['valence'] = (data['valence'] - data['valence'].min()) / (data['valence'].max() - data['valence'].min())
```

```python
# linear normailzation of numerical attributes
data['danceability'] = (data['danceability'] - data['danceability'].min()) / (data['danceability'].max() - data['danceability'].min())
data['loudness'] = (data['loudness'] - data['loudness'].min()) / (data['loudness'].max() - data['loudness'].min())
data['speechiness'] = (data['speechiness'] - data['speechiness'].min()) / (data['speechiness'].max() - data['speechiness'].min())
data['acousticness'] = (data['acousticness'] - data['acousticness'].min()) / (data['acousticness'].max() - data['acousticness'].min())
data['instrumentalness'] = (data['instrumentalness'] - data['instrumentalness'].min()) / (data['instrumentalness'].max() - data['instrumentalness'].min())
data['liveness'] = (data['liveness'] - data['liveness'].min()) / (data['liveness'].max() - data['liveness'].min())
data['tempo'] = (data['tempo'] - data['tempo'].min()) / (data['tempo'].max() - data['tempo'].min())
```

```python
# boolean to int
data['explicit'] = data['explicit'].astype(int)
```

```python
# label encoding of track_id, album_name, track_name, artists
data['track_id'] = data['track_id'].astype('category').cat.codes
data['album_name'] = data['album_name'].astype('category').cat.codes
data['track_name'] = data['track_name'].astype('category').cat.codes
data['artists'] = data['artists'].astype('category').cat.codes

# linear normalize these attributes
data['track_id'] = (data['track_id'] - data['track_id'].min()) / (data['track_id'].max() - data['track_id'].min())
data['album_name'] = (data['album_name'] - data['album_name'].min()) / (data['album_name'].max() - data['album_name'].min())
data['track_name'] = (data['track_name'] - data['track_name'].min()) / (data['track_name'].max() - data['track_name'].min())
data['artists'] = (data['artists'] - data['artists'].min()) / (data['artists'].max() - data['artists'].min())
```
